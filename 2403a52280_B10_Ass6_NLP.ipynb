Task-1

```python
import nltk
import re
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
import pandas as pd
```

```python
df = pd.read_excel('/content/LDA-Data.xlsx')
print(df)
```

 Cleaning and Lowercasing

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^"\w\s\[\]]', '', text)
    return text

df['cleaned_news'] = df['News'].apply(clean_text)
print(df)
```

## Tokenization

```python
nltk.download('punkt_tab')
print("Downloaded 'punkt_tab' NLTK resource.")
```

```python
from nltk.tokenize import word_tokenize

def tokenize_text(text):
    return word_tokenize(text)

df['tokenized_news'] = df['cleaned_news'].apply(tokenize_text)
print(df)
```

## Stopword Removal


```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]

df['news_without_stopwords'] = df['tokenized_news'].apply(remove_stopwords)
print(df)
```

## Lemmatization



```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def lemmatize_text(tokens):
    return [lemmatizer.lemmatize(word) for word in tokens]

df['lemmatized_news'] = df['news_without_stopwords'].apply(lemmatize_text)
print(df)
```

Rejoin

```python
def rejoin_text(tokens):
    return ' '.join(tokens)

df['preprocessed_news'] = df['lemmatized_news'].apply(rejoin_text)
print(df)
```

Bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the 'preprocessed_news' column
bow_matrix = vectorizer.fit_transform(df['preprocessed_news'])

# Display the Bag of Words matrix (sparse matrix) and feature names
print('Bag of Words Matrix (sparse):')
print(bow_matrix.toarray())
print('\nFeature Names (vocabulary):')
print(vectorizer.get_feature_names_out())
```

Bag of words as DataFrame

```python
bow_df = pd.DataFrame(bow_matrix.toarray(), columns=vectorizer.get_feature_names_out())
display(bow_df)
```

Apply LDA

```python
from sklearn.decomposition import LatentDirichletAllocation

# Instantiate LatentDirichletAllocation model
n_components = 2  # Number of topics
lda_model = LatentDirichletAllocation(n_components=n_components, random_state=42)

# Fit the LDA model to the Bag of Words matrix
lda_model.fit(bow_matrix)
print(f"LDA model fitted with {n_components} components.")
```

```python
feature_names = vectorizer.get_feature_names_out()
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))

no_top_words = 5
display_topics(lda_model, feature_names, no_top_words)
```

Topic modeling

```python
doc_topic_dist = lda_model.transform(bow_matrix)
print("Document-Topic Distribution (probabilities):")
print(doc_topic_dist)
```

```python
import numpy as np
dominant_topic = np.argmax(doc_topic_dist, axis=1)
print("Dominant topic for each document:")
print(dominant_topic)
```

```python
df_results = pd.DataFrame({
    'Original News': df['News'],
    'Dominant Topic': dominant_topic,
    'Topic 0 Probability': doc_topic_dist[:, 0],
    'Topic 1 Probability': doc_topic_dist[:, 1]
})
print("Topic Modeling Results per Document:")
print(df_results)
```

Text Preprocesing: Clean Text, Word Tokenization, stopword removal, lemma,Rejoin




```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')

def nltk_preprocessing_pipeline(text):
    # Initialize NLTK tools
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))

    # 1. Preprocess text (from previous steps)
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Remove mentions
    text = re.sub(r'@\w+', '', text)
    # Remove hashtags
    text = re.sub(r'#\w+', '', text)

    text = text.lower()  # Convert to lowercase

    # Remove emojis
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F1E0-\U0001F1FF"  # flags (iOS)
        "\U00002702-\U000027B0"
        "\U000024C2-\U0001F251"
        "]+", flags=re.UNICODE
    )
    text = emoji_pattern.sub(r'', text)

    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # Remove special characters
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces

    # 2. Word Tokenization
    tokenized_words = word_tokenize(text)

    # 3. Stopword Removal
    filtered_words = [word for word in tokenized_words if word not in stop_words]

    # 4. Lemmatization
    lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

    # 5. Rejoin words
    clean_summary = ' '.join(lemmatized_words)

    return clean_summary

print("NLTK preprocessing pipeline function created successfully!")
```

```python
data['clean_News'] = data['titles'].apply(nltk_preprocessing_pipeline)
print("\nComparison of previous clean_summaries and new clean_summaries_pipeline (first 5 rows):")
print(data[['clean_News']].head())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# Vectorize the cleaned summaries
count_vectorizer = CountVectorizer(max_df=0.95, min_df=1, stop_words='english')
doc_term_matrix = count_vectorizer.fit_transform(data['clean_News'])
```

BoW

```python
import pandas as pd
# Get feature (word) names
feature_names = count_vectorizer.get_feature_names_out()
# Convert sparse matrix to DataFrame
bow_df = pd.DataFrame(doc_term_matrix.toarray(),columns=feature_names)
# Display BoW matrix for top 10 documents
bow_top_10 = bow_df.head(10)
print(bow_top_10)
```

Apply LDA

```python
from sklearn.decomposition import LatentDirichletAllocation

# Initialize and fit LDA model
num_topics = 2
LDA = LatentDirichletAllocation(n_components=num_topics, random_state=42)
LDA.fit(doc_term_matrix)
```

```python
def display_topics(model, feature_names, num_top_words):
    for topic_idx in range(len(model.components_)):
        print(f"\nTopic {topic_idx}:")

        # Get word weights for this topic
        topic_weights = model.components_[topic_idx]

        # Get indices of words sorted by weight (descending)
        sorted_indices = topic_weights.argsort()[::-1]

        # Take top N words
        top_indices = sorted_indices[:num_top_words]

        # Print top words
        for idx in top_indices:
            print(feature_names[idx], end=" ")
        print()
```

Topic modeling

```python
# Display top words for each topic
num_top_words = 10
print(f"\nTop {num_top_words} words per topic:")
display_topics(LDA, count_vectorizer.get_feature_names_out(), num_top_words)
```

```python
# Assign topics to each document
document_topics = LDA.transform(doc_term_matrix)
data['topic'] = document_topics.argmax(axis=1)

print("\nDataFrame with assigned topics (first 5 rows):")
print(data[['clean_News', 'topic']].head())
```

NMF with Kaggle Data and BoW

```python
from sklearn.decomposition import NMF

nmf = NMF(n_components=2, random_state=42)
nmf.fit(doc_term_matrix)

display_topics(nmf, count_vectorizer.get_feature_names_out(), 10)
```

 NMF with Kaggle Data and TFIDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF

# Create TF-IDF matrix
tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=1, stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(data['clean_News'])

print("TF-IDF Matrix shape:", tfidf_matrix.shape)

# Initialize and fit NMF model
num_topics_nmf = 2
nmf_model_tfidf = NMF(n_components=num_topics_nmf, random_state=42)
nmf_model_tfidf.fit(tfidf_matrix)

print(f"\nNMF model fitted with {num_topics_nmf} components using TF-IDF.")

# Display topics
print("\nTop words per topic (NMF with TF-IDF):")
display_topics(nmf_model_tfidf, tfidf_vectorizer.get_feature_names_out(), 10)
```
